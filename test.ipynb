```python
import time
import random
import torchaudio
from glob import glob
import librosa
duration = 0
random.seed(10)
# files =glob("dataset/valid/*/*/*/vocals_128k.mp3")
# files =glob("dataset/test/*/*.wav")
files =glob("dataset/valid/*/*/*/vocals_sr16000.mp3")
random.shuffle(files)
for file in files[:120]:
    start = time.time()
    # torchaudio.info(file,backend="ffmpeg")
    torchaudio.load(file,backend="ffmpeg")
    # librosa.get_samplerate(file)
    librosa.load(file,sr=None,mono=True)
    duration += time.time() - start
duration
```
